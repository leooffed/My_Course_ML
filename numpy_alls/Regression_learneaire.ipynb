```python
import numpy as np
```

```python
# 1. Creation de donnees (ex: caracteristiques d'une maison: [surface de la maison, chambre])

maison_a = np.array([120, 3])
maison_b = np.array([150, 4])

# 2. Calcul de la similarite (Distance Eucludienne)
# C'est la base de l'algorithme K-Plus voisin (KNN)
regression = np.linalg.norm(maison_a - maison_b)

print(f"La distance entre les maisons est de: {regression}")
```
